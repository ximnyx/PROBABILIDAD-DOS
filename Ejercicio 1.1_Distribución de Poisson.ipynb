<div style="text-align: justify">

La distribución de Poisson debe su nombre al matemático francés **Siméon Denis Poisson** (1781–1840), quien la introdujo en 1837 dentro de su obra *Recherches sur la probabilité des jugements en matière criminelle et en matière civile*. En ese contexto, Poisson buscaba modelar la frecuencia de eventos raros, como errores judiciales, bajo ciertas condiciones de independencia y baja probabilidad individual.

Sin embargo, su uso práctico se consolidó años después gracias al estadístico inglés **Ladislaus Bortkiewicz**, quien a finales del siglo XIX analizó datos reales para validar el modelo. Su estudio más famoso examinó las muertes por patadas de caballo en el ejército prusiano, mostrando que estos eventos raros seguían aproximadamente una distribución de Poisson.

Desde entonces, esta distribución se ha convertido en una herramienta fundamental en probabilidad y estadística, especialmente para modelar conteos de eventos en intervalos de tiempo, espacio o volumen. Su relevancia se extiende a campos como la física, la biología, la ingeniería y las ciencias sociales, donde permite describir fenómenos aleatorios poco frecuentes con gran precisión.

$$

&=\frac{\lambda^x}{x!} \lim_{n \to \infty}

**3.** Escribe la función de masa de probabilidad de una variable aleatoria $X \sim \text{Poisson}(\lambda)$: $$\mathbb{P}(X=x)=\frac{e^{-\lambda}\lambda^x}{x!}, \qquad x \in \mathbb{N}_{0}$$

Explica por qué en este caso se habla de <i>función de masa de probabilidad</i> y no de <i>función de densidad</i>. Además, verifica que esta función satisface las condiciones básicas de una función de masa: no negatividad y normalización, es decir,
$$\sum_{x=0}^{\infty}\frac{e^{-\lambda}\lambda^x}{x!}=1.$$


</div>

La función de **masa de probabilidad de una variable aleatoria** $X \sim \text{Poisson}(\lambda)$ es:

$$
f(x)=P(X=x)=\frac{e^{-\lambda}\lambda^x}{x!}, \quad \text{para } x=0,1,2,\dots,\; \text{donde } \lambda> 0 \text{ es el parámetro de esta distribución.}
$$

### ¿Por qué es función de masa y no de densidad?

Se habla de **función de masa de probabilidad** porque $X$ es una variable aleatoria **discreta**, es decir, toma valores en un conjunto numerable ($0,1,2,\dots$).

En cambio, una **función de densidad** se utiliza cuando la variable es **continua**, y las probabilidades se obtienen mediante integrales.

---

### Verificación de las propiedades

#### 1. No negatividad

Para todo $x \in \mathbb{N}_0$:

$$
\frac{e^{-\lambda}\lambda^x}{x!} \geq 0
$$

ya que:
- $e^{-\lambda} > 0$
- $\lambda^x \geq 0$ si $\lambda > 0$
- $x! > 0$

Por lo tanto, $\mathbb{P}(X=x) \geq 0$.



#### 2. Normalización
$$
\begin{aligned}
\sum_{x=0}^{\infty}\frac{e^{-\lambda}\lambda^x}{x!}
&= e^{-\lambda} \sum_{x=0}^{\infty}\frac{\lambda^x}{x!}\\
&= e^{-\lambda} e^{\lambda}\\
&= e^{-\lambda+\lambda}\\
&= e^{0}\\
&= 1\\

\end{aligned}
$$

**4.** Define la función de distribución acumulada
      $$F(x)=\mathbb{P}(X\leq x)$$
      Escribe su expresión en el caso Poisson y explica por qué se trata de una función escalonada. Describe su comportamiento y señala cómo se relaciona con la función de masa de probabilidad.

</div>

La función de distribución acumulada se define como:

$$
F(x)=\mathbb{P}(X \leq x)
$$

Para $X \sim \text{Poisson}(\lambda)$, se tiene:

$$
F(x)=\mathbb{P}(X \le x)
=\sum_{k=0}^{x} \mathbb{P}(X = x)
= \sum_{k=0}^{x} \frac{e^{-\lambda}\lambda^k}{k!}
$$


---

### ¿Por qué es una función escalonada?

La función $F(x)$ es **escalonada** porque la variable $X$ es discreta.  
Esto significa que solo puede tomar valores enteros ($0,1,2,\dots$), por lo que:

- Entre dos enteros consecutivos, la función permanece constante.
- Solo “salta” en los valores enteros.


Sea $X \sim \text{Poisson}(\lambda)$ y su función de distribución acumulada:

$$
F(x)=\mathbb{P}(X \le x)
= \sum_{k=0}^{x} \frac{e^{-\lambda}\lambda^k}{k!}
$$



### Comportamiento de $F(x)$ y su relación con la función de masa de probabilidad

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, factorial

# Parámetro de Poisson
lam = 3

# Valores de x
x = np.arange(0, 15)

# Función de masa de probabilidad (PMF)
pmf = [(exp(-lam) * lam**k) / factorial(k) for k in x]

# Función de distribución acumulada (CDF)
cdf = np.cumsum(pmf)

# Gráfica
plt.figure(figsize=(8,5))

# PMF como barras
plt.bar(x, pmf, alpha=0.6, label='FMP: P(X = x) = f(x)')

# CDF como función escalonada
plt.step(x, cdf, where='post', label='FDA: P(X ≤ x) = F(x)')


for i in range(len(x)):
    plt.vlines(x[i], cdf[i]-pmf[i], cdf[i], linestyles='dashed', colors='red')


# Detalles
plt.xlabel('x')
plt.ylabel('Probabilidad')
plt.title('Relación entre FMP y FDA [X~Poisson(λ=3)]')
plt.legend()
plt.grid()
plt.show()
```

### ¿Entonces... cómo se relacionan $f(x)$ y $F(x)$?

La función de masa de probabilidad $f(x)=\mathbb{P}(X=x)$ determina los saltos de $F(x)$.

Es decir, **el tamaño del salto en cada entero es igual a la probabilidad en ese punto**.

**5.** Si $X \sim \text{Poisson}(\lambda)$, demuestra que
      $$\mathbb{E}[X]=\lambda \qquad \text{y} \qquad \mathrm{Var}(X)=\lambda$$
      Explica por qué esta distribución se caracteriza por tener media y varianza iguales.

$\underline{Demostración}$

$X \sim \text{Poisson}(\lambda)$ <br>
$\text{Por demostrar:}  \qquad ⓵ \quad \mathbb{E}[X]=\lambda \qquad \qquad ⓶ \quad \mathrm{Var}(X)=\lambda$



$$
\begin{aligned}

p(x) &= \frac{e^{-\lambda}\lambda^x}{x!} \qquad \qquad  \text{,} \qquad x = 0,1,\dots \\

⓵ \qquad \mathbb{E}[X] 
&= \sum_{x=0} x p(x) \\

&= \sum_{x=0}^{\infty} x \frac{e^{-\lambda}\lambda^x}{x!} \\

&= \sum_{x=1}^{\infty} \frac{x e^{-\lambda}\lambda^x}{x!} \\

&= \sum_{x=1}^{\infty} \frac{x e^{-\lambda}\lambda^x}{x(x-1)!} \\

&= e^{-\lambda} \sum_{x=1}^{\infty} \frac{\lambda^x}{(x-1)!} 
& \quad \text{// usando un cambio de variable} \\

&
& \begin{alignedat}{2}
\text{Sea } w &= x - 1 \;\Rightarrow\; &\quad&
\left\{
\begin{aligned}
x = 1 &\Rightarrow w = 0 \\
x \to \infty &\Rightarrow w \to \infty
\end{aligned}
\right. \\
&\Downarrow \\
&x = w + 1
\end{alignedat}
\\

&= e^{-\lambda} \sum_{w=0}^{\infty} \frac{\lambda^{w+1}}{w!} \\

&= e^{-\lambda} \sum_{w=0}^{\infty} \frac{\lambda^w \lambda}{w!} \\

&= e^{-\lambda} \lambda \sum_{w=0}^{\infty} \frac{\lambda^w}{w!} 
& \quad \text{// recordando que } e^x := \sum_{n=0}^{\infty} \frac{x^n}{n!} \\

&= e^{-\lambda} \lambda e^{\lambda} \\

&= e^{0} \lambda \\

&= \lambda \qquad\qquad \blacksquare

\end{aligned}
$$
